```python
# Defining function for finding minimun and maximun in dataframe
def min_max(data):
    min_=data.min()
    max_=data.max()
    return min_,max_
```

```python
# Columns used for calculation
cols = [c for c in df.columns if c not in ['Date', 'Hour','Minute','Day']]
```

### Finding operational temperature range

* To find the operational temperature range of each freezer, the minimum and maximum temperature of each freezer will be calculated. 

```python
# Printing minimum and maximum value for each freezer
for i in cols:
    range_=min_max(df[i])
    print("For ",i," Minimum Temperature : ",range_[0], ", Maximum Temperature : ",range_[1])
```

### Finding daily average of door open duration

* To find the daily average of door open duration, assumption is taken that the temperature of the freezer will increase only when the door of freezer will be opened and the temperature will decrease only when the door of freezer will be closed.

```python
# Defining function for calculating total duration of door open event 
def avg_opening(data):
    count=0
    for i in range(len(data)-1):
        if data[i] < data[i+1]:    # Checking if the temperature is greater than previous minute temperature
            count+=1               # If yes, increasing the minute counter by 1
    return count                   # Return total minute of door open event
```

```python
duration_dic={}   # This will contain the total open duration for each freezer 
for i in cols:
    count=avg_opening(df[i])  # Calculating open duration for each freezer
    # Since the data is for 16 days, so to find average duration i am dividing the data by 16
    print("Daily average of door open duration is :", count/16, "minutes for ", i)
    duration_dic[i]=count/16
```

### Classification of freezers based on operational health


* For this purpose,  freezers are categorised on the basis of their door open duration. The assumption taken here is that if the door is open it's mean it is operational in that period.

```python
for w in duration_dic:
    if (0 < duration_dic[w] < 120):  # This class will contains freezers which operational duration is 0 to 120 minutes
        print (w,": Class 1")
    elif (120 < duration_dic[w] < 240): # This class will contains freezers which operational duration is 120 to 240 minutes
        print(w,": Class 2")
    elif (240 < duration_dic[w] < 360): # This class will contains freezers which operational duration is 240 to 360 minutes
        print(w,": Class 3")
    elif (360 < duration_dic[w] < 480): # This class will contains freezers which operational duration is 360 to 480 minutes
        print(w,": Class 4")
    elif (480 < duration_dic[w] < 600): # This class will contains freezers which operational duration is 480 to 600 minutes
        print(w,": Class 5")
    elif (600 < duration_dic[w] < 720): # This class will contains freezers which operational duration is 600 to 720 minutes
        print(w,": Class 6")
    elif (720 < duration_dic[w] < 840): # This class will contains freezers which operational duration is 720 to 840 minutes
        print(w,": Class 7")
    elif (840 < duration_dic[w] < 1080): # This class will contains freezers which operational duration is 840 to 1080 minutes
        print(w,": Class 8")
```
