```python
import requests
from bs4 import BeautifulSoup
import re
from unidecode import unidecode

#Used headers/agent as the request timed out and asking for agent. Using following code you can fake the agent.
headers={'User-Agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:63.0) Gecko/20100101 Firefox/63.0'}
response = requests.get("https://www.hindustantimes.com/Error/Generic".encode('utf-8'),headers=headers)
text = response.text
print(text)
```

```python
soup= BeautifulSoup(text,"html.parser")
top_head1 =soup.find_all("div",attrs={"class": "media-heading headingfour"})
top_head2 =soup.find_all("div",attrs={"class": "media-heading headingfive"})
top_head3 =soup.find_all("div",attrs={"class": "headingfour"})
top_head4 =soup.find_all("div",attrs={"class": "headingfive"})
top_head5 =soup.find_all("div",attrs={"class": "random-heading"})
```

```python
top_head3[0]
```

```python
list1=[]
for i in range(len(top_head1)):
    regex = re.compile(r'[\n\r\t\u]')
    clean = regex.sub(" ", top_head1[i].text)
    #clean = unidecode(top_head1[i].text)
    a=clean.replace('\n', ' ')
    list1.append(a)
list1
```

```python
list2=[]
for i in range(len(top_head2)):
    clean = unidecode(top_head2[i].text)
    a=clean.replace('\n', ' ')
    list2.append(a)
```

```python
list3=[]
for i in range(len(top_head3)):
    clean = unidecode(top_head3[i].text)
    clean = regex.sub(" ", clean)
    a=clean.replace('\n', ' ')
    list3.append(a)
```

```python
list4=[]
for i in range(len(top_head4)):
    clean = unidecode(top_head4[i].text)
    clean = regex.sub(" ", clean)
    a=clean.replace('\n', ' ')
    list4.append(a)
```

```python
list5=[]
for i in range(len(top_head5)):
    clean = unidecode(top_head5[i].text)
    clean = regex.sub(" ", clean)
    a=clean.replace('\n', ' ')
    list5.append(a)
```

```python
final_list=list1+list2+list3+list4+list5
final_list
```

```python
import pandas
df = pandas.DataFrame(final_list)
df.to_csv("HtHeadlines.csv",index=False)
```
