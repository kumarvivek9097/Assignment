```python
# Import the required libraries
import pandas as pd
import numpy as np
import math
from datetime import datetime
```

```python
# Load the dataset
df=pd.read_csv('Temperature_Data.csv')
df.shape
```

```python
# Check the null values
df.isnull().sum()
```

```python
# Basic data view
df.head(50)
```

```python
# Rename first column
df.rename(columns={'0':'Date'},inplace=True)
```

```python
# Change string date to datetime format
df['Date']=pd.to_datetime(df['Date'], format='%Y-%m-%d %H:%M:%S')
```

```python
# Calculating day, hour and minute through datetime
df['Day']=df['Date'].dt.day
df['Hour']=df['Date'].dt.hour
df['Minute']=df['Date'].dt.minute
```

```python
# Grouping the data by the mean of each Hour
gk=df.groupby('Hour').mean()
```

```python
gk
```

```python
# Columns used for calculation
cols = [c for c in df.columns if c not in ['Date', 'Day','Hour','Minute']]
```

### Filling the missing values

There are various ways to fill the missing values. Some of the ways are:
* Replace the missing value with the next (or previous) value in the data frame
* Fill the missing values with the mean of the data
* Fill the missing values with the mode of the data

Since this is a time series data and the temperature largely depend upon freezer operational time, So I decided to fill the missing value with the average temperature of that particular hour.  

```python
# Filling the missing value
for i in cols:
    print(i)
    for j in range(len(df[i])):
        if(math.isnan(df[i][j])):      # Checking null values
            value=df['Hour'][j]       # 
            df[i][j]=gk[i][value]

# Here First I am seaching for index of missing row in each column, after getting index i am finding the hour in which
# that value is missing. After getting that hour I am filling the missing value by the mean temperature of that particular hour.
```

```python
# Check for null values
df.isnull().sum().sum()
```

```python
# Saving the data into dataframe
df.to_csv('new_data.csv',index=False)
```
